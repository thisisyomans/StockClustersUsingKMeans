# S&P 500 Stock Clusters Using K-Means

#### In this notebook, I download the pricing data for the S&P 500 stocks from 2017, calculate historic returns and volatility, and then use the K-Means clustering algorithm to divide the stocks into distinct groups based on the aforementioned returns and volatilities.

```python
from pylab import plot, show
from numpy import vstack, array
from numpy.random import rand
import numpy as np
from scipy.cluster.vq import kmeans, vq
import pandas as pd
import pandas_datareader as dr
from math import sqrt
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

##### In the cell below, I grab the S&P 500 stocks from Wikipedia and parse the data to create a list of tickers.

```python
SP500_URL = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
#read in the url and scrape ticker data
data_table = pd.read_html(SP500_URL)
tickers = data_table[0]['Symbol'].tolist()
```

##### In the cell below, I grab the ticker's price data using yahoo finance and create a dataframe with dates as row indices and tickers as columns.

```python
prices_df = pd.DataFrame(columns = tickers)

for ticker in tickers:
    try:
        prices_df[ticker] = dr.DataReader(ticker, 'yahoo', '01/01/2017')['Adj Close']
    except:
        print("Failed to read: {symbol}".format(symbol = ticker))
        del prices_df[ticker]
        
prices_df.head()
```

##### Now the I have the necessary data, I can actually being analysis. First, I need to decide how many clusters to separate the data into. Instead of choosing a random number of clusters, I used an "Elbow Curve" to highlight the relationship between how many clusters I choose and the Sum of Squared Estimate of Errors (SSE) resulting from using the number of clusters.

##### I then plot this relationship to help identify the optimal number of clusters. I would prefer less clusters, but also a lower SSE - a trade off that needs to be taken into account.

```python
#calculate average annual percentage return and volatility over a one year period
returns = prices_df.pct_change().mean() * 252
returns = pd.DataFrame(returns)
returns.columns = ['Returns']
returns['Volatility'] = prices_df.pct_change().std() * sqrt(252)

#returns.head()
#returns.isnull().any()
#format the data as a numpy array to feed into the K-Means algo
returns.drop('TT', inplace=True)
data = np.asarray([np.asarray(returns['Returns']), np.asarray(returns['Volatility'])]).T

X = data
distortions = []
for k in range(2, 20):
    k_means = KMeans(n_clusters=k)
    k_means.fit(X)
    distortions.append(k_means.inertia_)
    
fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), distortions)
plt.grid(True)
plt.title('Elbow curve')
```

##### In the above plot, we can see that once the number of clusters reaches 5, the reduction in SSE begins to slow quite a bit for each increase in cluster number. This led to me believe that the optimal number of clusters in this case was 5.

```python
#computing K-Means with K = 5 (5 clusters)
centroids, _ = kmeans(data, 5)
#assign each sample to a cluster
idx, _ = vq(data, centroids)

#some plotting using numpy's logical indexing
plot(data[idx==0,0], data[idx==0,1], 'ob', 
     data[idx==1,0], data[idx==1,1], 'oy', 
     data[idx==2,0], data[idx==2,1], 'or', 
     data[idx==3,0], data[idx==3,1], 'og', 
     data[idx==4,0], data[idx==4,1], 'om')
plot(centroids[:,0], centroids[:,1], 'sg', markersize=8)
show()
```

##### The above plot is a visual representation of the clusters. Originally, it was not so clean, as it had a major outlier (TT, as seen in the cell below) that was removed.
